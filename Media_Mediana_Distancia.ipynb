```python
# Dicionário para armazenar as distâncias entre média e mediana de cada crime
distancias_media_mediana = {}

# Calculando a distância entre média e mediana para cada crime
for crime in colunas_crimes:
    # Verifica se a coluna existe no DataFrame
    if crime in df.columns:
        media = df[crime].mean()
        mediana = df[crime].median()
        distancias_media_mediana[crime] = abs(media - mediana)
    else:
        print(f"Coluna '{crime}' não encontrada no DataFrame.")

# Exibindo as distâncias
for crime, distancia in distancias_media_mediana.items():
    print(f"Distância entre média e mediana de {crime}: {distancia:.2f}")
```

```python
def avaliar_confiabilidade_media(df, colunas, limite_baixo=10, limite_moderado=25):
    """
    Avalia a confiabilidade da média em relação à mediana para as colunas de um DataFrame.
    
    Parâmetros:
    - df (pd.DataFrame): DataFrame contendo os dados.
    - colunas (list): Lista de colunas a serem analisadas.
    - limite_baixo (float): Limite inferior percentual para a distância ser considerada confiável (padrão 10%).
    - limite_moderado (float): Limite superior percentual para influência moderada (padrão 25%).

    Retorna:
    - dict: Dicionário com as colunas e sua classificação de confiabilidade.
    """
    confiabilidade_resultados = {}

    for coluna in colunas:
        if coluna in df.columns:
            media = df[coluna].mean()
            mediana = df[coluna].median()
            distancia = abs(media - mediana)
            
            # Calcula a distância em relação ao percentual da mediana
            percentual_distancia = (distancia / mediana) * 100 if mediana != 0 else float('inf')
            
            # Classificação com base nos limites fornecidos
            if percentual_distancia <= limite_baixo:
                confiabilidade_resultados[coluna] = "Confiável"
            elif limite_baixo < percentual_distancia < limite_moderado:
                confiabilidade_resultados[coluna] = "Influência Moderada"
            else:
                confiabilidade_resultados[coluna] = "Alta Influência de Valores Extremos"
        else:
            print(f"Coluna '{coluna}' não encontrada no DataFrame.")
            confiabilidade_resultados[coluna] = None

    return confiabilidade_resultados

# Exemplo de uso
resultados_confiabilidade = avaliar_confiabilidade_media(df, colunas_crimes)

# Exibindo os resultados
for crime, status in resultados_confiabilidade.items():
    print(f"Confiabilidade da média de '{crime}': {status}")
```

```python
# Criar DataFrame para exibir confiabilidade dos crimes
df_confiabilidade = pd.DataFrame(
    resultados_confiabilidade.items(), 
    columns=['Crime', 'Confiabilidade']
)

# Exibir o DataFrame
display(df_confiabilidade)
```
